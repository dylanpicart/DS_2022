```python
# Create visual of survivors who are male and female
# Create a visual of survivors based on class
# Create visual of people who survived vs didn't

survived = titanic.loc[titanic['Survived'] == 1]
sex = survived['Sex'].value_counts()
sex
```

```python
survive = titanic['Survived'].value_counts()

axes = survive.plot.barh(y='Survived', color='purple', title='Died vs. Survived')
```

```python
sex.plot.bar(x='Sex', y='Survived')
plt.title('Survivors')
plt.xlabel('Sex')
plt.ylabel('Number of Survivors')
```
