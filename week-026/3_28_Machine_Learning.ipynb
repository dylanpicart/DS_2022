Loss Function - MSE/L2 Loss - Mean squared eror

Cost function J(theta0, theta1) = 1/m sum from i=1 to m of (y^i_true - y^i_predicted)^2

Minimize average squared difference between our mean value & our mean squared value

h_theta(x) = theta0 + theta1 * x : hypothesis
thea_0, theta_1 : parameters

Cost Function

Goal of MSE: To find a theta0 and theta1 that minimizes J(theta0, theta1)

```python
def naive_gradient_descent(x_set, y_set, alpha, tolerance, iterations, theta_1=0):
    list_of_thetas = [0]
    
    for i in range(0, iterations):
        # Represents our 1 / m component
        avg = 1 / (len(x_set))
        # Initialize the sum of the MSE
        sum_of_MSE = 0
        
        # Get the sum of the MSE
        for counter, value in enumerate(x_set):
            sum_of_MSE += (( (theta_1 * x_set[counter]) - y_set[counter]) * x_set[counter])
        # Calculate the new value of theta_1
        new_theta_1 = theta_1 - (alpha * avg * sum_of_MSE)
        # Add the new theta to the list of thetas
        list_of_thetas.append(new_theta_1)
        
        # Breaks the loop if convergence is reached
        if abs(theta_1 - new_theta_1) <= tolerance:
            return new_theta_1, list_of_thetas
        
        theta = new_theta_1
        
    return new_theta_1, list_of_thetas  
```

```python
def naive_gradient_descent_with_MSE(x, y, alpha=0.001, iterations=1000, tolerance=0.0001, theta_1=0):
    # 1/m
    avg = 1/(len(x))
    # Set up our for loop to handle the case where we fail to converge
    for i in range(iterations):
        # Start the process for calculating the Mean Squared Error
        MSE = 0
        
        for counter in range(len(x)):
            MSE += ((theta_1 * x[counter]) - y[counter]) * x[counter]
            
            # Update theta
            new_theta_1 = theta_1 - (alpha * avg * MSE)
            # If we converge, break the loop
            if abs(theta_1 - new_theta_1) < tolerance:
                return new_theta_1
            
        theta_1 = new_theta_1
            
    # If we don't converge, handle it
    print('Failed to converge')
```

```python
def gradient_descent_with_MSE(x, y, alpha=0.001, iterations=1000, tolerance=0.0001, theta_0=0, theta_1=0):
    
    avg = 1 / len(x)
    total_error = sum([(theta_0 + (theta_1*x[i]) - y[i])**2 for i in range(len(x))])
    
    for i in range(iterations):
        
        MSE_theta_0 = 0
        MSE_theta_1 = 0
        
        for counter in range(len(x)):
            MSE_theta_1 += ((theta_0 + theta_1 * x[counter]) - y[counter]) * x[counter]
            
        for counter in range(len(x)):
            MSE_theta_0 += ((theta_0 + theta_1 * x[counter]) - y[counter])
            
        new_theta_1 = theta_1 - ( alpha * avg * MSE_theta_1 )
        new_theta_0 = theta_0 - ( alpha * avg * MSE_theta_0 )
        
        new_total_error = sum([(new_theta_0 + (new_theta_1*x[i]) - y[i])**2 for i in range(len(x))])
        
        if abs(total_error-new_total_error) <= tolerance:
            return new_theta_0, new_theta_1
        total_error = new_total_error
        theta_0 = new_theta_0
        theta_1 = new_theta_1
    
    return new_theta_0, new_theta_1
```
